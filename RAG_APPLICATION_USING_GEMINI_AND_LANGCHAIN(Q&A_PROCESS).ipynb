```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(
    file_path = "/content/story of hyderabad.pdf" )
```

```python
doc=loader.load()
```

```python
doc
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)

final_doc=text_splitter.split_documents(doc)
final_doc[0]
```

```python

from google.colab import userdata
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
```

```python
GOOGLE_API_KEY
```

```python

embedding_model= GoogleGenerativeAIEmbeddings(google_api_key=GOOGLE_API_KEY, model="models/embedding-001")
```

```python
embedding_model
```

```python

## VectorStore Creation

vectorstore=FAISS.from_documents(final_doc,embedding_model)
```

```python
vectorstore
```

```python

##query using similarity search

query = "what is the capital of telangana"
relevant_doc = vectorstore.similarity_search(query)
relevant_doc
#relevant_doc[0].page_content
```

```python

retriever = vectorstore.as_retriever(search_type="similarity",search_kwargs={"k":5})
```

```python
retriever
```

```python
prompt_template="""
Use the following piece of context to answer the question asked.
Please try to provide the answer only based on the context.if the answer the dont know please give it has i dont the answer (naku telvadhu gurthu ledhu marchipoyina)

{context}
Question:{question}

Helpful Answers:
 """
```

```python
prompt=PromptTemplate.from_template(prompt_template)
```

```python
prompt
```

```python
retrievalQA=RetrievalQA.from_chain_type(
    llm=model,
    retriever=retriever,
    return_source_documents=True,
    chain_type_kwargs={"prompt":prompt}
)
```

```python
retrievalQA
```

```python

query = "what is the capital of chennai."
```

```python

# Call the QA chain with our query.
result = retrievalQA.invoke({"query": query})
print(result['result'])

```
